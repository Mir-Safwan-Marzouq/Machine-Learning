```python
import pandas as pd
import numpy as np
import math
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/homeprices2.csv")
df
```

**1 hot encoding for nominal data**

```python
#dummy variables need to be created

dummies = pd.get_dummies(df.town)
dummies
```

```python
# now concatenate this dummies dataframe with the original data frame

merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
# need to drop original town variable here
# drop one of the dummy variable cloumn because of dummy variable trap

final = merged.drop(['town', 'west windsor'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()  # Linear Regression Model Object created

```

```python
# X is nothing but all the columns except price column

X = final.drop(['price'],axis='columns')
X
```

```python
y = final.price
y
```

```python
# training your machine learning model using model.fit

model.fit(X,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(X,y) #accuracy check
```

**Using Sklearn one hot encoder**

```python
## using sklearn one hot encoder

df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() # label encoder class object is ready
```

```python
dfle=df
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
X = dfle[['town','area']].values
X
```

```python
y = dfle.price
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([("town", OneHotEncoder(), [0])], remainder = 'passthrough')
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]
X
```

```python
model.fit(X,y)
```

```python
model.predict([[1,0,2800]])
```
